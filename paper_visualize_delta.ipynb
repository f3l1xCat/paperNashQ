```python
# In[2]:

import numpy as np
import nashpy as ns
from itertools import permutations
import lemkeHowson1 as lemkeHowson
import matrix
import rational
import multiprocessing
import matplotlib.pyplot as plt
import csv
```

```python
plt.rc('font', family='Microsoft JhengHei')
res={}
with open('paper_delta', mode='r') as f:
    reader=csv.reader(f)
    for row in reader:
        if (len(row)==0):
            continue
        row[0]=float(row[0])
        for i in range(1, len(row)):
            row[i]=row[i].replace('(','')
            row[i]=row[i].replace(')','')
            row[i]=row[i].replace(',','')
            row[i]=tuple(map(float, row[i].split(' ')))
        # print(row)
        res[row[0]]=[row[i] for i in range(1, 4)]

alpha=[]
price=[]
eposide=[i for i in res]
for i in range(1,11):
    i=i/10
    avg_alpha_i=(res[i][0][0]+res[i][1][0]+res[i][2][0])/3
    avg_alpha_i=avg_alpha_i/100*3
    alpha.append(avg_alpha_i)
    avg_price_i=(res[i][0][1]+res[i][1][1]+res[i][2][1])/3
    avg_price_i=avg_price_i/100*3
    price.append(avg_price_i)
    # print(avg_alpha_i, avg_price_i)

# 創建圖表
fig, ax = plt.subplots(figsize=(4, 3))

# 繪製曲線
ax.plot(eposide, alpha, label=None, color='blue', linestyle='-')
ax.plot(eposide, price, label=None, color='red', linestyle='--')

# 添加標註和箭頭
ax.annotate(r'$\alpha$(廣告數量)',
            xy=(0.3, alpha[2]), xycoords='data', # xy 是標註箭頭指向的位置。
            xytext=(0.2, 0.7), textcoords='data', # xytext 是標註文字的位置。
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=.5"))

ax.annotate(r'$p$(平台收費價格)',
            xy=(0.6, price[5]), xycoords='data',
            xytext=(0.7, 0.7), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=.2"))

# 設置標籤和標題
ax.set_xlabel(r'殘餘搜尋量 $\delta$')
ax.set_ylabel(r'平台決策變數')
ax.set_title(r'不同$\delta$下的平台決策變數變化圖')
ax.set_xlim(0.1, 1)

# 顯示圖例
# ax.legend()

# 顯示圖表
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 假設這些數據是你所需要的
w = np.linspace(0.1, 0.63, 500)
pi_S_I = 0.1 * np.sin(2 * np.pi * w) + 0.04
pi_S_II = 0.05 * np.sin(2 * np.pi * w - np.pi / 2) + 0.06
pi_S_III = 0.03 * np.sin(2 * np.pi * w - np.pi) + 0.03

# 創建圖表
fig, ax = plt.subplots()

# 繪製曲線
ax.plot(w, pi_S_I, label=r'$\pi_{S,I}$', color='black', linestyle='-')
ax.plot(w, pi_S_II, label=r'$\pi_{S,II}$', color='black', linestyle='--')
ax.plot(w, pi_S_III, label=r'$\pi_{S,III}$', color='black', linestyle='-.')

# 添加標註和箭頭
ax.annotate(r'$\omega$',
            xy=(0.31, pi_S_II[155]), xycoords='data',
            xytext=(0.4, 0.08), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=.5"))

ax.annotate(r'$\pi_{S,I}$',
            xy=(0.22, pi_S_I[70]), xycoords='data',
            xytext=(0.15, 0.04), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=.2"))

ax.annotate(r'$\pi_{S,II}$',
            xy=(0.31, pi_S_II[155]), xycoords='data',
            xytext=(0.35, 0.09), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=.2"))

ax.annotate(r'$\pi_{S,III}$',
            xy=(0.52, pi_S_III[255]), xycoords='data',
            xytext=(0.45, 0.02), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=.2"))

# 添加垂直虛線
ax.axvline(x=0.1, color='black', linestyle=':')
ax.axvline(x=0.31, color='black', linestyle=':')
ax.axvline(x=0.52, color='black', linestyle=':')

# 設置標籤和標題
ax.set_xlabel(r'Fixed reward $w$')
ax.set_ylabel(r'Contest sponsor\'s payoff')
ax.set_title('Comparison of Contest Sponsor\'s Payoffs')

# 顯示圖例
ax.legend()

# 顯示圖表
plt.show()
```
